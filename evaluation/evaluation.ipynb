```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the time data from CSV
time_data = pd.read_csv("time.csv", index_col=0)

# Compute the average time spent per task
avg_time_per_task = time_data.mean(axis=1)

# Load the interaction data from CSV
interaction_data = pd.read_csv("num_of_interactions.csv", index_col=0)

# Compute the average number of interactions
avg_interactions = interaction_data.mean(axis=1)


# Define x-axis labels and positions
x_labels = list(avg_time_per_task.index)
x_positions = np.arange(len(x_labels))

# Interaction data should align with the corresponding tasks
interaction_x = [x_labels.index("Looking at the visualizations"), x_labels.index("Reading the suggestions")]  # Indices for the two points
interaction_y = [avg_interactions["Switch between visualization_script"],
                 avg_interactions["Switch between suggestion_script"]]

# Create the figure
fig, ax1 = plt.subplots(figsize=(9, 6))
ax1.bar(x_positions, avg_time_per_task.values, color="#FFC2A1", label="Time Spent (seconds)")
ax1.set_ylabel("Time Spent (seconds)", color="#E57360")
ax1.tick_params(axis="y", labelcolor="#E57360")
ax1.set_ylim(0, 230)

# Create second y-axis for number of interactions
ax2 = ax1.twinx()
ax2.scatter(interaction_x, interaction_y, color="#005079", s=100, label="Number of Interactions", marker="o")  # Use dots with custom color
ax2.set_ylabel("Number of Interactions", color="#005079")
ax2.tick_params(axis="y", labelcolor="#005079")
ax2.set_ylim(0, 7)

# Set x-axis labels
ax1.set_xticks(x_positions)
ax1.set_xticklabels(x_labels, rotation=15)

# Set integer intervals for number of interactions axis
ax2.set_yticks([1, 2, 3, 4, 5, 6, 7])

# Title & Legends
plt.title("Time Spent and Number of Interactions Per Task (Combined)")
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Show the plot
plt.show()
```
